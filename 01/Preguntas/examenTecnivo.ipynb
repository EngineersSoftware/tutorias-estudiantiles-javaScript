### 9. Escribe el c√≥digo para crear un servidor en Node.js que devuelva:
- 'Inicio' si se accede a `/` con GET
- 'Usuario registrado' si se accede a `/usuario` con POST
- 'Ruta no v√°lida' para cualquier otra combinaci√≥n



````python
const http = require('http');

const server = http.createServer((req, res) => {
  if (req.url === '/' && req.method === 'GET') {
    res.end('Inicio');
  } else if (req.url === '/usuario' && req.method === 'POST') {
    res.end('Usuario registrado');
  } else {
    res.end('Ruta no v√°lida');
  }
});

server.listen(3000, () => {
  console.log('Servidor corriendo en http://localhost:3000');
});
````
